```python
X_mean = X.mean()
X_std = X.std()

X_standardized = (X - X_mean) / X_std
```

```python
X_standardized
```

```python
X_min = X.min()
X_max = X.max()

X_norm = (X - X_min) / (X_max - X_min)
```

```python
X_norm
```

```python
m = np.random.randn()
c = np.random.randn()

print("Randomly initialized slope:", m)
print("Randomly initialized intercept:", c)
```

```python
def predict(X, m, c):
    y_pred = m * X + c
    return y_pred
```

```python
# Make predictions using initial values
y_pred = predict(X_norm, m, c)
```

```python
# Display first 5 predictions
print("First 5 predicted values:")
print(y_pred[:5])
```

```python
# Mean Squared Error (cost function)

def compute_cost(y, y_pred):
    n = len(y)
    cost = (1 / (2 * n)) * np.sum((y_pred - y) ** 2)
    return cost
```

```python
# Compute initial cost
cost = compute_cost(y, y_pred)

print("Initial Cost (MSE):", cost)
```

```python
# Compute predictions for given parameters
y_pred = predict(X_norm, m, c)
```

```python
# Compute cost for given m and c
cost = compute_cost(y, y_pred)

print("Cost for given parameters:")
print("Slope (m):", m)
print("Intercept (c):", c)
print("Cost (MSE):", cost)
```

```python
# Given parameters
m = 0.0
c = 0.0

# Prediction
y_pred = predict(X_norm, m, c)

# Cost calculation
cost = compute_cost(y, y_pred)

print("Cost for m =", m, "and c =", c, "is:", cost)
```

```python
#Define Learning Rate
```

```python
learning_rate = 0.001   # step size for gradient descent
epochs = 2000          # number of training iterations

print("Learning Rate:", learning_rate)
print("Number of Epochs:", epochs)
```

```python
# Number of training examples
n = len(y)
```

```python
# Lists to store cost values (optional, for plotting)
cost_history = []
```

```python
# Gradient Descent loop
for epoch in range(epochs):

    # Step 1: Prediction
    y_pred = predict(X_norm, m, c)

    # Step 2: Compute gradients
    dm = (1 / n) * np.sum((y_pred - y) * X_norm)
    dc = (1 / n) * np.sum(y_pred - y)

    # Step 3: Update parameters
    m = m - learning_rate * dm
    c = c - learning_rate * dc

    # Step 4: Compute and store cost
    cost = compute_cost(y, y_pred)
    cost_history.append(cost)

    # Print cost occasionally
    if epoch % 100 == 0:
        print(f"Epoch {epoch}: Cost = {cost}")
```

```python
# Store final trained parameters

final_slope = m
final_intercept = c

print("Final Model Parameters:")
print("Slope (m):", final_slope)
print("Intercept (c):", final_intercept)
```

```python
def predict_salary(years_experience, m, c, X_norm):


    # Predict salary
    salary_pred = m * X_norm + c
    return salary_pred
```

```python
# Example: Predict salary for 5 years of experience
years = 5

predicted_salary = predict_salary(
    years,
    final_slope,
    final_intercept,
    X_norm
)

print(f"Predicted salary for {years} years of experience:")
print(predicted_salary[0][0])
```

```python
# Predict outputs using trained model
y_pred_final = predict(X_norm, final_slope, final_intercept)

# Calculate Mean Squared Error
mse = np.mean((y - y_pred_final) ** 2)

print("Mean Squared Error (MSE):", mse)
```
